### 1. Merging two dataframes using merge()

```python
# Creating 2 dataframes 


df1 = pd.DataFrame({"J.No": ["183","44","18","07","12"],
                    'Name': ["Virat_Kohli","Micheal_Hussey","Adam_Gilchrist","M.S.Dhoni","Sourav_Ganguly"],
                    'Batting_Style': ["RH","LH","LH","RH","LF"],
                    'Highest_Score': ["183","142","183","183","183"],
                   })
df2 = pd.DataFrame({'Average':["52.02","44.05","42.09","49.09","42.92"],
                     'Nationality':["Indian","Australian","Australian","Indian","Indian"],
                     "J.No": ["18","423","02","48","198"]})

dm = pd.merge(df1,df2,how = 'outer',on = 'J.No')
dm
```

###  2. Changing the OBJECT datatype to NUMERIC datatype:

### 4. Finding duplicated values in a DataFrame

```python
df1 = pd.DataFrame({"J.No": ["183","44","18","07","12"],
                    'Name': ["Virat_Kohli","Micheal_Hussey","Adam_Gilchrist","M.S.Dhoni","Sourav_Ganguly"],
                    'Batting_Style': ["RH","LH","LH","RH","LF"],
                    'Highest_Score': ["183","142","183","183","183"],
                   })
df2 = pd.DataFrame({'Average':["52.02","44.05","42.09","49.09","42.92"],
                     'Nationality':["Indian","Australian","Australian","Indian","Indian"],
                     "J.No": ["18","423","02","48","198"]})

print(df1.duplicated('Batting_Style'))
print(df2.duplicated('Nationality'))
```

### 5. Using REPLACE() in pandas to replace a single value in a dataframe

```python
dataframe = pd.DataFrame({
   "Phones":["IPhone","Samsung","MI","Realme"],
    "Laptops":["HP","Dell","Lenovo","MacBookPro"]
})
print(dataframe)

#Using the replace() function
dataframe.replace({'Phones':"IPhone","Laptops":"MacBookPro"},"Apple")
```

### 8.Text Manupulation using pandas using 'split' and 'apply'

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_months_data['Purchase_city'] = all_months_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
all_months_data.head()
```

### 9. Creating date and time format from the data (Using date_range, .dt.year etc)

```python
datedf = pd.DataFrame()
datedf['date'] = pd.date_range('24/5/2020',periods = 24, freq = 'H')


datedf['year'] = datedf['date'].dt.year 
datedf['month'] = datedf['date'].dt.month 
datedf['day'] = datedf['date'].dt.day 
datedf['hour'] = datedf['date'].dt.hour 
datedf['minute'] = datedf['date'].dt.minute 

datedf.head(10)
```
